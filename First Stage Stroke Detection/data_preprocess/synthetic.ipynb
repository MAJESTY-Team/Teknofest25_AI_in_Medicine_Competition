```python
import os
import shutil
from pathlib import Path

IMAGE_EXTENSIONS = {'.png'}

def read_images(root_dir):
    root = Path(root_dir)
    for path in root.rglob('*'):
        if path.suffix.lower() in IMAGE_EXTENSIONS:
            yield path


# Source and destination directories
train_dir = "../trainingDatav3/train"
bt_sentetik_dir = "../BT-Sentetik Part1/"
dest_dir = "../trainingDatav3.1/train/"

# Create destination directory if it doesn't exist
os.makedirs(dest_dir, exist_ok=True)

# Read all files in the train directory
for filename in read_images(train_dir):
    # Create the corresponding synthetic filename
    synthetic_filename = f"1-{filename.name}"

    synthetic_path = os.path.join(bt_sentetik_dir, filename.parent.name, synthetic_filename)

    # # Check if the synthetic file exists
    # print(synthetic_path)
    if os.path.exists(synthetic_path):
        # Copy the synthetic file to destination
        dest_path = os.path.join(dest_dir, filename.parent.name, synthetic_filename)
        shutil.copy2(synthetic_path, dest_path)
        print(f"Copied: {synthetic_filename}")
    else:
        print(f"Not found: {synthetic_filename}")

print("Copy operation completed.")
```
