```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11m-cls")  # load a pretrained model (recommended for training)
# Train the model

# Train the model
results = model.train(
    data='../trainingDatav3.1/',
    epochs=300,
    name="yolo11m-cls_v3.1",
    batch=64,
    erasing=0.0,
    weight_decay=0.0005,
    dropout=0.25,
    patience=50,)
```

```python
def calculate_f1_score(tp, fp, fn):
    precision = tp / (tp + fp) if tp + fp > 0 else 0
    recall = tp / (tp + fn) if tp + fn > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
    return f1
```
