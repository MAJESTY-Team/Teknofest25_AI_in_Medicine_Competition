```python
import pandas as pd
import os
```

```python
os.getcwd()
```

```python
directories=["İnme Yok Veri Seti","İskemi Veri Seti","Kanama Veri Seti"]
names=["İnme Yok","İskemi","Kanama"]
```

```python
images={dir:os.listdir(dir) for dir in directories}
len(images)
```

```python
for dir,name in zip(directories,names):
    for image in images[dir]:
        if image.endswith(".png"):
            os.rename(os.path.join(dir, image), os.path.join("Data/images", f"{image[:-4]}_{name}.png"))
```

```python
ims=os.listdir("Data/images")
```

```python
def get_label(name):
    return name[6:-4]
```

```python
df=pd.DataFrame(ims,columns=["imagelabels"])
# df.iloc[0].apply(get_label, axis=1)
# part the IDs and labels
df["ids"]=df["imagelabels"].apply(lambda x: x.split("_")[0])
df["imagelabels"]=df["imagelabels"].apply(lambda x: x.split("_")[1][:-4])

# df["imagelabels"]=df["imagelabels"].apply(lambda x: x.split("_"),expand=True)
# df.iloc[:,0]=df.iloc[:,0].apply(lambda x: x.split("_"),expand=True)
y=df["imagelabels"].to_numpy()
X=df["ids"].to_numpy()
```

```python
df
```

```python
# First split: 90% train+val, 10% test (stratified)
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Check class distribution
from collections import Counter
print("Train:", Counter(y_train))
print("Val:", Counter(y_val))
# print("Test:", Counter(y_test))
```

```python

save_dir = 'ClassificationData/val'
os.makedirs(save_dir, exist_ok=True)


for name in names:
    os.makedirs(os.path.join(save_dir, name), exist_ok=True)

for id, label in zip(X_val, y_val):
    try:
        os.rename(os.path.join(f"{label} Veri Seti", f"{id}.png"), os.path.join(save_dir, label, f"{id}_{label}.png"))
    except:
        print(f"File not found: {os.path.join(f'{label} Veri Seti', f'{id}.png')}")
        continue
```
