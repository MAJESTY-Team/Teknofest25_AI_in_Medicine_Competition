```python
from ultralytics import YOLO
model = YOLO("../TrainingCodes/runs/classify/yolo11n-cls_topluv2_/weights/best.pt") 

read_dir="../Yolo/Toplu/val"
output_dir="./val_results"
```

```python
import shutil
from pathlib import Path
import os
IMAGE_EXTENSIONS = {'.png'}

def read_images(root_dir):
    root = Path(root_dir)
    for path in root.rglob('*'):
        if path.suffix.lower() in IMAGE_EXTENSIONS:
            yield path

# Usage
for path in read_images(read_dir):
    if(str(path).find("CT")!=-1):
        continue

    label=path.stem.split("_")[-1]
    # name=f"{caseNum}_"+path.name
    result=model.predict(source=path, save=False)
    result=result[0]
    pred=result.probs.top1
    if result.names[pred].lower() == label.lower():
        dst=os.path.join(output_dir, "T")
    else:
        dst=os.path.join(output_dir, "F")
    os.makedirs(dst, exist_ok=True)
    name=path.name
    shutil.copyfile(path, os.path.join(dst, name))
    print(f"True label: {label}, Predicted: {result.names[pred]}")
    print(f"Image saved to {os.path.join(dst, name)}")
```

```python
import os
from PIL import Image, ImageEnhance, ImageOps
import random

# 1) Configuration: point these to your folders
input_dir = 'Assistance_required_data'    # Folder containing original images
output_dir = '../Yolo/Toplu/train'  # Folder to save augmented images
aug_per_image = 3            # How many augmented copies per original

# Ensure output folder exists
os.makedirs(output_dir, exist_ok=True)

def random_flip(img):
    """Random horizontal and/or vertical flip."""
    if random.choice([True, False]):
        img = ImageOps.mirror(img)
    if random.choice([True, False]):
        img = ImageOps.flip(img)
    return img

def random_rotation(img):
    """Rotate by a random angle between -30 and +30 degrees."""
    angle = random.uniform(-40, 40)
    return img.rotate(angle, expand=True)

def random_color_jitter(img):
    """Adjust brightness, contrast, and color randomly."""
    # Brightness
    enhancer = ImageEnhance.Brightness(img)
    img = enhancer.enhance(random.uniform(0.7, 1.3))

    # Contrast
    enhancer = ImageEnhance.Contrast(img)
    img = enhancer.enhance(random.uniform(0.7, 1.3))

    # # Color/Saturation
    # enhancer = ImageEnhance.Color(img)
    # img = enhancer.enhance(random.uniform(0.8, 1.2))

    return img

# def random_crop(img):
#     """
#     Randomly crop 80100% of the original, then resize back to original size.
#     This simulates a zoom-and-translate effect.
#     """
#     width, height = img.size
#     scale = random.uniform(0.8, 1.0)
#     new_w = int(width * scale)
#     new_h = int(height * scale)
#     left = random.randint(0, width - new_w)
#     top = random.randint(0, height - new_h)
#     cropped = img.crop((left, top, left + new_w, top + new_h))
#     return cropped.resize((width, height), Image.ANTIALIAS)

def augment_image(image_path, output_base, count):
    """
    Apply a sequence of random augmentations and save the results.
    Each run picks a random combination of flip, rotate, color jitter, crop.
    """
    img = Image.open(image_path).convert('RGB')
    base_name = os.path.splitext(os.path.basename(image_path))[0]
    label=base_name.split("_")[-1]
    output_base = os.path.join(output_base, label)
    for i in range(count):
        aug = img.copy()

        # Chain a few random transforms:
        aug = random_flip(aug)
        aug = random_rotation(aug)
        aug = random_color_jitter(aug)

        # Save out as JPEG (you can switch format/quality if desired)
        save_name = f"aug.{i+1}.{base_name}.png"
        aug.save(os.path.join(output_base, save_name), quality=100)

# Loop through each image and produce augmented variants
for image_file in read_images(input_dir):
    image_path = image_file
    augment_image(image_path, output_dir, aug_per_image)
    print(f"Processed {image_path}")

# print(f"Processed {len(image_files)} images, each with {aug_per_image} variants.")
```
