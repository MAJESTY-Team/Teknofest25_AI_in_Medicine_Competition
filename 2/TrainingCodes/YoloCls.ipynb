```python
from ultralytics import YOLO

# Load a model
model = YOLO("runs\classify\yolo11m-cls_topluv2_/weights/last.pt")  # load a pretrained model (recommended for training)
# Train the model
results = model.train(data='../Yolo/Toplu', epochs=100,name="yolo11m-cls_topluv2_", batch=64, erasing=0.0)
```

```python
def calculate_f1_score(tp, fp, fn):
    precision = tp / (tp + fp) if tp + fp > 0 else 0
    recall = tp / (tp + fn) if tp + fn > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
    return f1
```
