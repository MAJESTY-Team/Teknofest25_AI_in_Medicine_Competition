```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11m-cls")  # load a pretrained model (recommended for training)
# Train the model
results = model.train(data='./newTraining', epochs=250,name="final_yolo11m_", batch=64, erasing=0.0)
```

```python
def calculate_f1_score(tp, fp, fn):
    precision = tp / (tp + fp) if tp + fp > 0 else 0
    recall = tp / (tp + fn) if tp + fn > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
    return f1
```
