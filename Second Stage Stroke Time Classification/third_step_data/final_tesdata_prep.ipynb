```python
def read_dicom(path):

    ds = pydicom.dcmread(path)
    # 2. Convert to pixel array
    arr=ds.pixel_array
    # 3. (Optional) Apply rescale slope/intercept
    slope = getattr(ds, 'RescaleSlope', 1.0)
    intercept = getattr(ds, 'RescaleIntercept', 0.0)
    arr = arr * slope + intercept

    # 4. Windowing (convert to 0–255)
    #    If the DICOM has WindowCenter/Width tags, use them:
    if ('WindowCenter' in ds) and ('WindowWidth' in ds):
        center = ds.WindowCenter
        width = ds.WindowWidth
        # Handle multi-valued tags
        if isinstance(center, pydicom.multival.MultiValue):
            center = center[0]
            width = width[0]
        low = center - width/2
        high = center + width/2
        arr = np.clip(arr, low, high)
    arr = (arr - arr.min()) / (arr.max() - arr.min())  # normalize 0–1
    arr = (arr * 255).astype(np.uint8)                  # scale to 0–255
    return arr
```

```python
from pathlib import Path
from PIL import Image
import os
save_dir="png_test"

IMAGE_EXTENSIONS = {'.dcm'}

def read_images(root_dir):
    root = Path(root_dir)
    for path in root.rglob('*'):
        if path.suffix.lower() in IMAGE_EXTENSIONS:
            img = read_dicom(path)
            yield path, img

os.makedirs(save_dir, exist_ok=True)

for path, img in read_images(wdir):

    img = Image.fromarray(img)
    img.save(os.path.join(save_dir, "Vaka_" + path.parent.parent.name+"_"+str(path.name)[:-4] + ".png"))
    print(f"{path}: size={img.size}")
```

```python
import pandas as pd
import os

pathA='./preprocessed_third_step_data'

if(not os.path.exists(pathA)):
    assert False, f"Path {pathA} does not exist."
    
df=pd.DataFrame(os.listdir(pathA))
# Rename the original column
df = df.rename(columns={0: 'filename'})

# # Split the filenames by "_" and create new columns
df[["vaka",'case_number', 'slice_info', "label"]] = df['filename'].str.split('_', n=4, expand=True)


df
```

```python
# First split: 90% train+val, 10% test (stratified)
from sklearn.model_selection import train_test_split

X=df["slice_info"].to_numpy()
y=df["label"].to_numpy()

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.6, stratify=y, random_state=42
)

# Check class distribution
from collections import Counter
print("Train:", Counter(y_train))
print("Val:", Counter(y_val))
# print("Test:", Counter(y_test))
```

```python
parent_dir='./preprocessed_third_step_data'
dst_dir="../Yolo/Topluv2"
import shutil

for im in X_train:
    fileName=df[df["slice_info"]==im]["filename"].values[0]
    label=df[df["slice_info"]==im]["label"].values[0][:-4]
    src=os.path.join(parent_dir, fileName)
    os.makedirs(os.path.join(dst_dir,"train/"+ label), exist_ok=True)
    dst=os.path.join(dst_dir,"train/"+ label, fileName)
    shutil.copyfile(src, dst)


for im in X_val:
    fileName=df[df["slice_info"]==im]["filename"].values[0]
    label=df[df["slice_info"]==im]["label"].values[0][:-4]
    src=os.path.join(parent_dir, fileName)
    os.makedirs(os.path.join(dst_dir,"val/"+ label), exist_ok=True)
    dst=os.path.join(dst_dir,"val/"+ label, fileName)
    shutil.copyfile(src, dst)
```
