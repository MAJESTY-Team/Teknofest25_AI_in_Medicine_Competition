```python
from ultralytics import YOLO

# Load a model
model = YOLO("./runs/classify/yolo11m-cls_init/weights/best.pt")  # load a pretrained model (recommended for training)
# Train the model
results = model.train(
    data='../trainingData_nonLeakage/',
    epochs=150,
    name="yolo11n-cls_v2_extra_finetune",
    lr0=1e-4,  # Adjust learning rate
    batch=64,
    weight_decay=0.0005,  # Add weight decay
    dropout=0.25,          # Add dropout to the classifier
)
```

```python
def calculate_f1_score(tp, fp, fn):
    precision = tp / (tp + fp) if tp + fp > 0 else 0
    recall = tp / (tp + fn) if tp + fn > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
    return f1
```
